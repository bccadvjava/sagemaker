# category classification
### load sklearn toy dataset on breast cancer#

```python
import sklearn.datasets as dt
bunch = dt.load_breast_cancer()
print (bunch.DESCR)
```

```python
bunch
```

# load pandas, and load data into a dataframe

```python
import pandas as pd
dataDF = pd.DataFrame(bunch.data, columns = bunch.feature_names)
dataDF
```

# scale all the columns on a range of zero to ten

```python
import sklearn.preprocessing as prep
scalar = prep.MinMaxScaler((0,10), False)
scaledData = scalar.fit_transform(dataDF)
scaledDF = pd.DataFrame(scaledData, columns = bunch.feature_names)
scaledDF
```

```python
scaledDF.describe()
```

# do a single feature analsys
### create dataframes having only one input feature

```python
targetDF = pd.DataFrame(bunch.target, columns=['Outcome'])
oneFacDF = scaledDF[['mean radius']]
oneFacGraphDF = oneFacDF.join(targetDF)
oneFacGraphDF
```

## create plot to assess efficiency of logistic regression

```python
import matplotlib.pyplot as plt
plt.plot(oneFacGraphDF['mean radius'], oneFacGraphDF.Outcome, 'gx ')
plt.show()
```

## create plot t0 assess efficiency of SVM

```python
import numpy as np
colors = ['bo ', 'rx ']
for i in [0,1]:
    x = oneFacGraphDF.loc[(oneFacGraphDF.Outcome==i)]['mean radius']
    y = np.zeros(x.shape) + ((i+1)/20)
    plt.ylim(0,1.0)
    plt.plot(x, y, colors[i])
plt.show()
```

## run KNN, SVM, logistic regression classifiers, get prediction scores

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import ShuffleSplit
from sklearn.linear_model import LogisticRegression
KNNscore = 0
SVMscore = 0
LRscore = 0
#oneFacDF = scaledDF[['mean radius']]
rs = ShuffleSplit(n_splits=5, test_size=.25, random_state=0)
for trainIndex, testIndex in rs.split(oneFacDF):
    trainDFx = oneFacDF.iloc[trainIndex]
    testDFx = oneFacDF.iloc[testIndex]
    trainDFy = bunch.target[trainIndex]
    testDFy = bunch.target[testIndex]
    
    neigh = KNeighborsClassifier()
    neigh.fit(trainDFx, trainDFy)
    KNNscore += neigh.score(testDFx, testDFy)
    
    svm = SVC(gamma='scale')
    svm.fit(trainDFx, trainDFy)
    SVMscore += svm.score(testDFx, testDFy)
    
    lr = LogisticRegression(solver='lbfgs')
    lr.fit(trainDFx, trainDFy)
    LRscore += lr.score(testDFx, testDFy)
    
print ("Average prediction score for KNN:", format(100 * KNNscore / 5, '.2f'))
print ("Average prediction score for SVM:", format(100 * SVMscore / 5, '.2f'))
print ("Average prediction score for LR :", format(100 * LRscore / 5, '.2f'))
#print ("There are", svm.n_support_, 'support vectors in SVM')
#svm.support_vectors_
print ('SVM: coef:', lr.coef_, 'intercept', lr.intercept_)
```

## try a prediction

```python
classes = ['Malignant', 'Benign' ]
pred = neigh.predict([[0]])
print ('Prediction for x=0:', classes[pred[0]])
pred = neigh.predict([[5]])
print ('Prediction for x=5:', classes[pred[0]])
pred = neigh.predict([[9]])
print ('Prediction for x=9:', classes[pred[0]])
```

# run the classifiers for the complete feature set

```python
KNNscore = 0
SVMscore = 0
LRscore = 0
rs = ShuffleSplit(n_splits=5, test_size=.25, random_state=0)
for trainIndex, testIndex in rs.split(scaledDF):
    testDFx = scaledDF.iloc[testIndex]
    trainDFx = scaledDF.iloc[trainIndex]
    trainDFy = bunch.target[trainIndex]
    testDFy = bunch.target[testIndex]
    
    neigh = KNeighborsClassifier()
    neigh.fit(trainDFx, trainDFy)
    KNNscore += neigh.score(testDFx, testDFy)
    
    svm = SVC(gamma='scale')
    svm.fit(trainDFx, trainDFy)
    SVMscore += svm.score(testDFx, testDFy)
    
    lr = LogisticRegression(solver='lbfgs', max_iter=1000)
    lr.fit(trainDFx, trainDFy)
    LRscore += lr.score(testDFx, testDFy)
    
print ("Average prediction score for KNN:", format(100 * KNNscore / 5, '.2f'))
print ("Average prediction score for SVM:", format(100 * SVMscore / 5, '.2f'))
print ("Average prediction score for LR :", format(100 * LRscore / 5, '.2f'))
```

# let's try a PCA analysis on the complete feature set

```python
from sklearn.decomposition import PCA
import numpy as np
pca = PCA()
pca.fit(scaledData)
pcaData = pca.transform(scaledData)
percentVar = np.round(pca.explained_variance_ratio_ * 100, decimals=1)
percentVar
#pcaData.shape
```

```python
labels = []
for i in range(0,len(percentVar)):
    labels.append('PC' + str(i))
plt.bar(labels[:8], percentVar[:8])
plt.title('Scree plot')
plt.show()
```

```python
pcaDF = pd.DataFrame(pcaData, columns=labels)
pcaDF["Outcome"] = bunch.target
pcaDF
```

## chart the first two principal components

```python
for i in range(len(colors)):
    clusterMembers = np.array(pcaDF.loc[(pcaDF.Outcome==i)][["PC0" , "PC1"]])
    plt.plot(clusterMembers[:,0], clusterMembers[:,1], colors[i], label=classes[i])
    print('class', i, 'has:', clusterMembers.shape)
plt.xlabel('PC0 - {0}%'.format(percentVar[0]))
plt.ylabel('PC1 - {0}%'.format(percentVar[1]))
plt.legend()
plt.show()
```

```python
PCwtg = pd.DataFrame(pca.components_[:2].T, index = bunch.feature_names, columns = ['PC0', 'PC1'])
PCwtg = PCwtg.sort_values('PC0', ascending=0)
PCwtg
```

```python
redDF = pcaDF[['PC0', 'PC1', 'Outcome']]
redDF.describe()
```

## run the classifiers against the principal components

```python
KNNscore = 0
SVMscore = 0
LRscore = 0
redDF = pcaDF[['PC0', 'PC1', 'PC2', 'PC3', 'PC4']]
rs = ShuffleSplit(n_splits=5, test_size=.25, random_state=0)
for trainIndex, testIndex in rs.split(redDF):
    trainDFx = redDF.iloc[trainIndex]
    testDFx = redDF.iloc[testIndex]
    trainDFy = bunch.target[trainIndex]
    testDFy = bunch.target[testIndex]
    
    neigh = KNeighborsClassifier()
    neigh.fit(trainDFx, trainDFy)
    KNNscore += neigh.score(testDFx, testDFy)
    
    svm = SVC(gamma='scale')
    svm.fit(trainDFx, trainDFy)
    SVMscore += svm.score(testDFx, testDFy)
    
    lr = LogisticRegression(solver='lbfgs')
    lr.fit(trainDFx, trainDFy)
    LRscore += lr.score(testDFx, testDFy)
    
print ("Average prediction score for KNN:", format(100 * KNNscore / 5, '.2f'))
print ("Average prediction score for SVM:", format(100 * SVMscore / 5, '.2f'))
print ("Average prediction score for LR :", format(100 * LRscore / 5, '.2f'))
```
